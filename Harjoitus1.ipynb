```python
import pandas as pd
```

```python
df = pd.read_excel('https://taanila.fi/data1.xlsx')
```

```python
df[(df["koulutus"]!=1) & (df["palkka"]<=1700)]
```

```python
df.nsmallest(3, "palkka")
```

```python
df.nlargest(5, "ikä")
```

```python
df.describe()
```

```python
#Frekvenssitaulukko Crosstab!
df1 = pd.crosstab(df["koulutus"], "f")
df1
```

```python
df1.columns.name=""
df1
```

```python
#Tehdään lista koulutusmuuttujaan ja sen jälkeen sijoitetaan lista dataframen indeksiin
koulutus = ["Peruskoulu", "2.aste", "Korkeakoulu", "Ylempi korkeakoulu"]
df1.index=koulutus
#Huomaa, että df1 on muuttuja, joka sisältää frekvenssitaulukon.
df1
```

```python
#Taulukkoon voidaan luoda sarake ja siihen voidaan viitata
n = df1["f"].sum() #Havaintojen lukumäärä
#Lisätään sarake %, joka on f sarakkeen rivin % lukema. Rivin/kokonaislukumäärä*100
df1["%"] = df1["f"]/n*100
df1
```

```python
#Lisätään uusi rivi .loc komennolla
df1.loc["Yhteensä"] = df1.sum()
df1
```

```python
df2 = pd.crosstab(df["koulutus"], df["sukup"]) #Voidaan tehdä myös ristiintaulukointi
df2
```

```python
sukupuoli = ["Mies", "Nainen"]
df2.index = koulutus #Menee indeksi sarakkeeseen
df2.columns = sukupuoli #Menee sarakeotsikoihin
df2
```

```python
df2.index.name = "Koulutus"
df2.columns.name = "Sukupuoli"
df2
```

```python
df = df.rename(columns={"sukup":"sukupuoli", "palveluv":"palveluvuodet"})
df.columns
```

```python
df2 = df2.reset_index() #Indeksi siirretaan pois indeksistä ja tilalle laitetaan oletus indeksit 0,1,2...
df2
```

```python
df2 = df2.set_index("Koulutus")
df2
```

```python
#Kuinka monta 2. asteen koulutuksen suorittanutta miestä on?
df2.loc["2.aste", "Mies"]
```

```python
#Pelkkä : tarkoittaa kaikkia rivejä ja kaikkia sarakkeita
df2.loc[:, "Nainen"]
```

```python
df2.loc[:, "Nainen"].to_frame() #Tämä tekee tuloksesta dataframen
```

```python
#Muodostetaan luokka CUT funktiolla. Bins parametrin arvoksi tulee lista
bins = [19, 29, 39, 49, 59, 69]

df["ikäluokka"] = pd.cut(df["ikä"], bins=bins) #Huomioi, että arvojen tulee nousta tietyssä järjestyksessä. Ei voi olla esim 19-29 ja 29-32, 39-49
df
```

```python
#Tehdään frekvenssitaulukko ikäluokista
df3 = pd.crosstab(df["ikäluokka"], "ikä")
df3.columns.name = ""
df3
```

Ryhmittely

```python
#Luodaan GroupBy ryhmittely. Funktiolle pitää antaa parametrinä sen muuttujan nimi, jonka mukaan ryhmitellään
df4 = df.groupby("koulutus")["palkka"].describe()
df4.index = koulutus
df4
```

Luettelo muistiossa käytetyistä funktioista ja ominaisuuksista

Huomaa kuitenkin, että Pythonin muuttujien nimiä ei kirjoitetan heittomerkkien väliin (esimerkiksi df, n jne.)
