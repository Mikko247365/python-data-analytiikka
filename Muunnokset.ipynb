Datan muuntaminen

```python
import pandas as pd

df = pd.read_excel('https://taanila.fi/data1.xlsx')
df
```

```python
df.columns
```

```python
#Muutetaan sarakkeiden nimet sukup = sukupuoli, palveluv = palveluvuodet perhe = perhesuhde
df.columns = ['nro', 'sukupuoli', 'ikä', 'perhesuhde', 'koulutus', 'palveluvuodet', 'palkka',
       'johto', 'työtov', 'työymp', 'palkkat', 'työteht', 'työterv', 'lomaosa',
       'kuntosa', 'hieroja']
df.columns
```

```python
#Myös raname() funtiolla pystyy muuttamaan. Tarvitaan sanakirja {}
df = df.rename(columns={"sukupuoli":"sukup", "perhesuhde":"perhe", "palveluvuodet":"palvelup"})
df.columns
```

Sarakkeiden poistaminen

```python
#Drop funktiolla
#Derop postaa oletuksena rivejä AXIS=0, sarakkeet ovat AXIS=1
df1 = df.drop(["työterv", "lomaosa", "kuntosa", "hieroja"], axis=1)
df1
```

Indeksi

```python
#Muutuja nro siirretään indeksiin set_index() funtiolla
df1 = df1.set_index("nro")
df1
```

```python
#reset_index muuttaa sen takaisin tavalliseksi sarakkeeksi
df1 = df1.reset_index()
df1
```

Virheellisten arvojen korjaaminen

```python
#Korjataan virheellisen iän ikään 42 indeksinumeron 47 rivillä
df1.loc[47, "ikä"] = 42
#Katsotaan onko indeksinumeron 47 ikä rivillä 42
df1.iloc[40:50]
```

Uusien sarakkeiden luonti

```python
#Luodaan sukupuoli_str, jossa sukupuolet ovat tekstinä
#Käytetään sanakirjaa
df1["sukupuoli_str"] = df1["sukup"].replace({1:"Mies", 2:"Nainen"})
df1
```

```python
#Luodaan luokkarajat ikäluokille. HUOMAA, ETTÄ PITÄÄ OLLA SAMANLAISET HYPPÄYKSET
bins = [19, 29, 39, 49, 59, 69]

#Luodaan uusisarake, jossa iät on luokiteltu ikäluokkiin
df1["ikäluokka"] = pd.cut(df1["ikä"], bins=bins)

df1
```

```python
#Lasketaan vastaajan "kokonaistyytyväisyyden" keskiarvona tyytyväiksyyksistä eri osa-alueilla
#axis=1 tarkoittaa laskemista sivusuunnassa! Rivien keskiarvo
df1["tyytyväisyys"] = df1[["johto", "työtov", "työymp", "palkkat", "työteht"]].mean(axis=1)

df1[["johto", "työtov", "työymp", "palkkat", "työteht", "tyytyväisyys"]]
```

```python
#Uusi sarake, jossa on laskettu käytettyjen etuuksien lukumäärä .count
df["käyttö"] = df[["työterv", "lomaosa", "kuntosa", "hieroja"]].count(axis=1)
df[["työterv", "lomaosa", "kuntosa", "hieroja", "käyttö"]]
```

Puuttuvia arvoja sisältävien rivien poistaminen

```python
#Poistetaan rivit, joilta puuttuu näiden sarakkeiden arvot

df2 = df1.dropna(subset=["koulutus", "palvelup", "työtov"])
df2
```
