```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("Opinnäytetyökysely.xlsx")
colum = df["Sain riittävästi ohjausta"]

ax = sns.histplot(df, x=colum, bins=5)
bins_centers = [patch.get_x() + patch.get_width() / 2 for patch in ax.patches]
plt.xticks(bins_centers, labels=[f'{int(x)}' for x in bins_centers])
plt.show()
```

```python
x = df["Pystyin itse vaikuttamaan opinnäytetyöni ohjaajan valintaan"]
y = df["Sain riittävästi ohjausta"]

sns.scatterplot(df, x=x, y=y)
plt.xticks(range(1,6,1))
plt.show()
```

Molempien muuttujien arvot ovat samalla asteikolla 1-5, joten havaintoja esimerkiksi akselilla x=1 ja y=2 voi olla esimerkiksi 10kpl, mutta se näyttäydy kaaviossa määrällisenä. Tämän takia data näyttää harvalta ja informaatio katoaa.

```python
df2 = df.dropna(subset=["Thesis grade"])
x2 = df2["Opinnäytetyön tekemisaika työviikkoina (40 h) aihekuvauksen tekemisestä työn valmistumiseen:työviikkoa"]
y2 = df2["Thesis grade"]

sns.scatterplot(df2, x=x2, y=y2)
plt.xlabel("Aika viikkoina")
plt.ylabel("Arvosana")
plt.title("Arvosanan riippuvuus tekoajasta")
plt.show()
```
