```python
import pandas as pd
```

```python
df = pd.read_excel("Opinnäytetyökysely.xlsx")
df
```

```python
var = ["Opinnäytetyön tekemiseen kulunut aika ensimmäisistä aihekaavailuista työn valmistumiseen:kuukautta"]

df1 = df.groupby("Opiskeluala")[var].describe()
columns = ["Lukumäärä", "Keskiarvo", "Keskihajonta", "Pienin", "Alaneljännes", "Mediaani", "Yläneljännes", "Suurin"]
df1.columns=columns
df1.style.format("{:.1f}")
```

```python
df2 = df.groupby("Sukupuoli")[var].describe()
columns = ["Lukumäärä", "Keskiarvo", "Keskihajonta", "Pienin", "Alaneljännes", "Mediaani", "Yläneljännes", "Suurin"]
df2.columns=columns
df2.style.format("{:.1f}")
```

```python
df3 = df.pivot_table(values="Sain riittävästi ohjausta", index="Opiskeluala", columns="Sukupuoli", margins=True, margins_name="Kaikki")
df3.index.name=""
df3.columns.name=""
df3.style.format("{:.1f}")
```

```python
df4 = df.pivot_table(values=["Luotin ohjaajani neuvoihin", "Ohjaajaani oli helppo lähestyä"], index="Opiskeluala", columns="Sukupuoli")
df4.style.format("{:.1f}")
```
