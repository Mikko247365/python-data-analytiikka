```python
import pandas as pd
from scipy.stats import pearsonr
```

```python
df = pd.read_excel("Opinnäytetyökysely.xlsx")
df
```

```python
df1 = df["Hankin itse aktiivisesti tietoa työni aiheesta"].corr(df["Tutkimusaiheeni kiinnosti minua"])
r = df1**2

print(f"Pearsonin korrelaatiokerroin on:{df1:.02f}\nSelitysaste on: {r:.02f}")
```

```python
df1 = df["Sain riittävästi ohjausta"].corr(df["Opinnäytetyön tekemiseen kulunut aika ensimmäisistä aihekaavailuista työn valmistumiseen:kuukautta"])
r = df1**2

print(f"Pearsonin korrelaatiokerroin on:{df1:.02f}\nSelitysaste on: {r:.02f}")
```

```python
df["Uusi keskiarvo"]=df[["Ohjaajani panos tuki työtäni",
                        "Saamani ohjaus oli asiantuntevaa",
                        "Saamani ohjaus oli motivoivaa",
                        "Työni ohjaaja vastasi nopeasti tiedusteluihini",
                        "Ohjaustilanteet eivät tuntuneet minusta pelottavilta",
                        "Ohjaajaani oli helppo lähestyä",
                        "Luotin ohjaajani neuvoihin"]].mean(axis=1)
```

```python
df2 = df["Pystyin itse vaikuttamaan opinnäytetyöni ohjaajan valintaan"].corr(df["Uusi keskiarvo"])
r = df2**2

print(f"Pearsonin korrelaatiokerroin on:{df2:.02f}\nSelitysaste on: {r:.02f}")
```

```python
target = "Olin innostunut opinnäytetyötä tehdessäni"
cols = ["Ohjaajan tuki opinnäytetyön eri vaiheissa:aiheen valinnassa",
        "Ohjaajan tuki opinnäytetyön eri vaiheissa:tutkimuskysymysten tai kehittämistehtävän rajauksessa",
        "Ohjaajan tuki opinnäytetyön eri vaiheissa:menetelmien valinnassa",
        "Ohjaajan tuki opinnäytetyön eri vaiheissa:aineiston analyysissa",
        "Ohjaajan tuki opinnäytetyön eri vaiheissa:johtopäätösten ja yhteenvedon tekemisessa",
        "Ohjaajan tuki opinnäytetyön eri vaiheissa:raportoinnissa"]
print("Pearsonin testi Olin innostunut opinnäytetyötä tehdessäni:")
for col in cols:
    df3 = df[[target, col]].dropna()
    r, p = pearsonr(df3[target], df3[col])
    print(f'{col:<25} r = {r:.3f}, p = {p:.3f}')
```
