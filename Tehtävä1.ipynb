Tehdään frekvenssitaulukko

```python

df1 = pd.crosstab(df["Kuinka löysit aiheesi?"], "f")

n = df1["f"].sum()
df1["%"] = (df1["f"]/n)*100
df1.columns.name=""
df1.loc["Yhteensä"] = df1.sum()
df1.style.format({"f":"{:.0f}", "%":"{:.1f} %"})
```
