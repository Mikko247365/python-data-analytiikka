```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("Opinnäytetyökysely.xlsx")
df
```

```python
df2 = pd.crosstab(df["Opiskeluala"], columns=df["Sukupuoli"])
df2.columns.name=""
df2.index=["Kulttuuri", "Liiketalous", "Tekniikka"]
df2
```

```python
df3 = df.groupby("Sukupuoli")["Opiskeluala"].value_counts(normalize=True).unstack("Sukupuoli")*100
n_men = (df["Sukupuoli"]=="Mies").sum()
n_women = (df["Sukupuoli"]=="Nainen").sum()
columns = [f"Mies, n = {n_men}", f"Nainen, n = {n_women}"]
df3.columns=columns
df3.index=["Kulttuuri", "Liiketalous", "Tekniikka"]
df3.style.format("{:.1f} %", na_rep="-", decimal=".")
```

```python
df4 = pd.crosstab(df["Sukupuoli"], df["Opiskeluala"], normalize="index")*100
df4.index=[f'Mies, n = {n_men}', f'Nainen, n = {n_women}']
df4.columns.name=""
df4.style.format("{:.1f} %", na_rep="-", decimal=".")
```

```python
df5 = df.groupby("Sukupuoli")["Oliko työ parityö?"].sum().to_frame("Oliko työ parityö?")
df5.index=[f'Miehet, n = {n_men}', f'Naiset, n = {n_women}']
df5
```
